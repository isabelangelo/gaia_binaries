```python
%pylab inline
import pandas as pd
```

```python
training_labels = pd.read_csv('data/label_dataframes/training_labels.csv')
test_labels = pd.read_csv('data/label_dataframes/test_labels.csv')
```

```python
figure(figsize=(9,5))
rcParams['font.size']=10
ctr = 'orange'
cte = 'darkolivegreen'
subplot(231);xlabel(r'$T_{eff}$ (K)');ylabel('number of stars')
hist(training_labels.galah_teff, color=ctr)
hist(test_labels.galah_teff, color=cte)
text(6100,1500,'training set', color=ctr)
text(6100,1300,'test set', color=cte)
subplot(232);xlabel('logg (dex)');ylabel('number of stars')
hist(training_labels.galah_logg, color=ctr)
hist(test_labels.galah_logg, color=cte)
subplot(233);xlabel('[Fe/H] (dex)');ylabel('number of stars')
hist(training_labels.galah_feh, color=ctr)
hist(test_labels.galah_feh, color=cte)
subplot(234);xlabel(r'[$\alpha$/Fe] (dex)');ylabel('number of stars')
hist(training_labels.galah_alpha, color=ctr)
hist(test_labels.galah_alpha, color=cte)
subplot(235);xlabel(r'$v_{broad}$ (km/s)');ylabel('number of stars')
hist(training_labels.galah_vbroad, color=ctr)
hist(test_labels.galah_vbroad, color=cte)
subplots_adjust(hspace=0.3, wspace=0.4)
```
