```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

training_set_df = pd.read_csv('../train_cannon_model/data_files/cannon_training_set.csv')
test_set_df = pd.read_csv('../train_cannon_model/data_files/cannon_test_set.csv')
```

```python
# I need the same bins as before
Teff_bins = np.arange(3000,10000,300)
logg_bins = np.arange(0,7,0.5)
fe_h_bins = np.arange(-3,2,0.25)
gmag_bins = np.arange(0,25,1)
bprp_bins = np.arange(0,8,0.5)
vbroad_bins = np.arange(0,300,3)
training_color='#FFC300'
test_color='#BEE37A'

plt.figure(figsize=(15,8))
plt.subplot(231)
plt.hist(training_set_df.galah_teff, 
         bins=Teff_bins, histtype='step', lw=3,color=training_color)
plt.hist(test_set_df.galah_teff, 
         bins=Teff_bins, histtype='step', lw=3,color=test_color)
plt.text(6200,1000,'training set', color=training_color, fontsize=20)
plt.text(6200,800,'     test set', color=test_color, fontsize=20)
plt.xlabel('teff')

plt.subplot(232)
plt.hist(training_set_df.galah_logg, 
         bins=logg_bins, histtype='step', lw=3,color=training_color)
plt.hist(test_set_df.galah_logg, 
         bins=logg_bins, histtype='step', lw=3,color=test_color)
plt.xlabel('logg')

plt.subplot(233)
plt.hist(training_set_df.galah_feh, 
         bins=fe_h_bins, histtype='step', lw=3,color=training_color)
plt.hist(test_set_df.galah_feh, 
         bins=fe_h_bins, histtype='step', lw=3,color=test_color)
plt.xlabel('fe/h')

plt.subplot(234)
plt.hist(training_set_df.galah_alpha_fe, 
         bins=fe_h_bins, histtype='step', lw=3,color=training_color)
plt.hist(test_set_df.galah_feh, 
         bins=fe_h_bins, histtype='step', lw=3,color=test_color)
plt.xlabel('alpha/fe')

plt.subplot(235)
plt.hist(training_set_df.galah_vbroad, 
         bins=vbroad_bins, histtype='step', lw=3,color=training_color)
plt.hist(test_set_df.galah_vbroad, 
         bins=vbroad_bins, histtype='step', lw=3,color=test_color)
plt.xlabel('vbroad')
plt.xlim(0,100)
```
