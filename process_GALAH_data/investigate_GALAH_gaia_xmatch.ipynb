```python
from astropy.table import Table
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
GALAH_xmatch = Table.read('./GALAH_data_tables/GALAH_DR3_VAC_GaiaEDR3_v2.fits', format='fits').to_pandas()
GALAH_star_labels = Table.read('./GALAH_data_tables/GALAH_DR3_main_allstar_v2.fits', format='fits').to_pandas()
```

```python
df = pd.merge(GALAH_xmatch, GALAH_star_labels, on='sobject_id')
df
```

```python
df_no_duplicates = df.drop_duplicates(subset='designation', keep='last')
df_no_duplicates
```

```python
# note: when I drop duplicates but keep the last one, there is 588465 targets
# when I drop all duplicates, there is 588464 targets
# this means that there are 107 targets that have a duplicated Gaia designation
588571 - 588464
```

```python
# next: find all the Gaia designations that occur more than once
designations = df.designation.to_numpy()
for designation in designations:
    length = len(np.where(designations==designation)[0])
    if length!=1:
        print(designation, length)
```
