```python
import pandas as pd
```

```python
cks_label_df = pd.read_csv('./data/cks_labels.csv')
training_label_df = pd.read_csv('../gaia_binaries/data/galah_label_dataframes/training_labels.csv')
```

```python
%pylab inline

figure(figsize=(7,7))
subplot(221)
hist(cks_label_df.cks_steff, label='CKS sample with\nRVS spectra')
hist(training_label_df.galah_teff, alpha=0.5, label='Cannon training set')
xlabel('Teff (K)');legend(loc='upper left', frameon=False, fontsize=10)

subplot(222)
hist(cks_label_df.cks_slogg)
hist(training_label_df.galah_logg, alpha=0.5)
xlabel('logg (dex)')

subplot(223)
hist(cks_label_df.cks_smet)
hist(training_label_df.galah_feh, alpha=0.5)
xlabel('[Fe/H] (dex)')

subplot(224)
hist(cks_label_df.cks_svsini)
hist(training_label_df.galah_vbroad, alpha=0.5)
xlabel('vbroad (km/s)')

# this says that I need to remove stars with vsini<minimum training set value
# and try again!
```
